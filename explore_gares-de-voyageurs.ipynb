```python
import pandas as pd
```

```python
filepath = "./raw/gares-de-voyageurs.csv"
```

```python
df = pd.read_csv(filepath,
                 header=0,
                 sep=";"
                 )
```

```python
df.head(3)
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df_nan = df.loc[df.isna().any(axis=1)]
df_nan
```

```python
df.loc[df['Nom'] == 'Neuilly Porte Maillot RER E', 'Position géographique'] = df.loc[df['Nom'] == 'Neuilly Porte Maillot RER E', 'Position géographique'].fillna('48.877983, 2.284516')
```

```python
df.dropna(subset=['Position géographique'], inplace=True)
```

```python
df.duplicated().sum()
```

```python
df.to_csv("./cleaned/gares-de-voyageurs.csv", sep=";", index=False)
```

```python
import folium
```

```python
m = folium.Map(location=[48.8566, 2.3522], zoom_start=6)  # Centrée sur Paris
```

```python
for index, row in df.iterrows():
    lat, lon = map(float, row['Position géographique'].split(', '))
    folium.Marker(
        location=[lat, lon],
        popup=row['Nom'],
        tooltip=row['Nom']
    ).add_to(m)
```

```python
m.save('map.html')
```
