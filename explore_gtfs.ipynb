### stop_times

```python
df_stop_times = pd.read_csv(stop_times)
df_stop_times.head()

# trip_id: Identifiant du voyage
# arrival_time: Heure d'arrivée
# departure_time: Heure de départ
# stop_id: Identifiant de l'arrêt
# stop_sequence: Ordre de l'arrêt dans le voyage
# stop_headsign: Destination affichée à l'arrêt
# pickup_type: Type de montée autorisée (0: Autorisée, 1: Non autorisée, 2: Demande d'autorisation)
# drop_off_type: Type de descente autorisée (0: Autorisée, 1: Non autorisée, 2: Demande d'autorisation)
# shape_dist_traveled: Distance parcourue depuis le début du voyage
```
